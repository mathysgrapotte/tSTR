```python
blank_dict = {'lr':0,
             'batch_size':0,
             'filter1':0,
             'filter2':0,
             'filter3':0,
             'noyau1':0,
             'noyau2':0,
             'noyau3':0,
             'linear1':0,
             'linear2':0}

config = copy.deepcopy(blank_dict)
path_save = "configs_CNN/"

for l in lines:
    
    if l[0:17] == "Best trial config":
        data = l.split('{')[1].split('}')[0].split(",")
        config['lr'] = float(data[0].split(" ")[-1])
        config['batch_size'] = int(data[1].split(" ")[-1])
        config['filter1'] = int(data[2].split(" ")[-1])
        config['filter2'] = int(data[3].split(" ")[-1])
        config['filter3'] = int(data[4].split(" ")[-1])
        config['noyau1'] = int(data[5].split(" ")[-1])
        config['noyau2'] = int(data[6].split(" ")[-1])
        config['noyau3'] = int(data[7].split(" ")[-1])
        config['linear1'] = int(data[8].split(" ")[-1])
        config['linear2'] = int(data[9].split(" ")[-1])
        
    if l[0] == '/':
        str_class = l.split('/')[-1].split('_')[0]
        path_save_class = path_save + "config_" + str_class + ".npy"
        np.save(path_save_class, config)
        config = copy.deepcopy(blank_dict)
    
```
